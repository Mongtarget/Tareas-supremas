```python
#.1 LISTA CIRCULAR

class Nodo:
    def __init__(self, dato):
        self.dato = dato
        self.siguiente = None


class ListaCircular:
    def __init__(self):
        self.cabeza = None

    def es_vacia(self):
        return self.cabeza is None

    def longitud(self):
        if self.es_vacia():
            return 0
        nodoActual = self.cabeza
        cantidad = 0
        while True:
            cantidad += 1
            nodoActual = nodoActual.siguiente
            if nodoActual == self.cabeza:
                break
        return cantidad

    def imprimirLista(self):
        if self.es_vacia():
            print("La lista está vacía")
            return
        nodoActual = self.cabeza
        print(nodoActual.dato)
        while nodoActual.siguiente != self.cabeza:
            nodoActual = nodoActual.siguiente
            print(nodoActual.dato)

    def agregarNodoListaCircular(self, dato):
        nodo = Nodo(dato)
        if self.es_vacia():
            self.cabeza = nodo
            nodo.siguiente = self.cabeza
        else:
            nodoActual = self.cabeza
            while nodoActual.siguiente != self.cabeza:
                nodoActual = nodoActual.siguiente
            nodoActual.siguiente = nodo
            nodo.siguiente = self.cabeza
            self.cabeza = nodo

    def agregarNodoUltimoListaCircular(self, dato):
        nodo = Nodo(dato)
        if self.es_vacia():
            self.cabeza = nodo
            nodo.siguiente = self.cabeza
        else:
            nodoActual = self.cabeza
            while nodoActual.siguiente != self.cabeza:
                nodoActual = nodoActual.siguiente
            nodoActual.siguiente = nodo
            nodo.siguiente = self.cabeza

    def agregarNodoIesimaListaCircular(self, indice, dato):
        if indice < 0 or indice > self.longitud():
            print("No existe tal posición")
            return False
        elif indice == 0:
            self.agregarNodoListaCircular(dato)
        else:
            nodo = Nodo(dato)
            nodoActual = self.cabeza
            cantidad = 0
            while cantidad < indice:
                anterior = nodoActual
                nodoActual = nodoActual.siguiente
                cantidad += 1
            anterior.siguiente = nodo
            nodo.siguiente = nodoActual

    def eliminarNodoEspecifico(self, dato):
        if self.es_vacia():
            return
        elif dato == self.cabeza.dato:
            nodoActual = self.cabeza
            while nodoActual.siguiente != self.cabeza:
                nodoActual = nodoActual.siguiente
            nodoActual.siguiente = self.cabeza.siguiente
            self.cabeza = self.cabeza.siguiente
        else:
            nodoActual = self.cabeza
            anterior = None
            while nodoActual.dato != dato:
                anterior = nodoActual
                nodoActual = nodoActual.siguiente
                if nodoActual == self.cabeza:
                    print("Elemento no encontrado")
                    return
            anterior.siguiente = nodoActual.siguiente


# Ejemplo de uso:
if __name__ == "__main__":
    lista = ListaCircular()
    lista.agregarNodoUltimoListaCircular(10)
    lista.agregarNodoUltimoListaCircular(20)
    lista.agregarNodoUltimoListaCircular(30)
    lista.imprimirLista()
    print("Longitud:", lista.longitud())
    lista.eliminarNodoEspecifico(20)
    lista.imprimirLista()
```

```python
#2 LISTA DOBLEMENTE ENLAZADA

class Nodo:
    def __init__(self, dato):
        self.dato = dato
        self.siguiente = None
        self.anterior = None


class DoublyLinkedList:
    def __init__(self):
        self.inicio = None

    def insertEmptyList(self, data):
        if self.inicio is None:
            newNode = Nodo(data)
            self.inicio = newNode
        else:
            print("La lista no está vacía")

    def insertBeginList(self, data):
        if self.inicio is None:
            newNode = Nodo(data)
            self.inicio = newNode
            print("Nodo insertado al inicio (lista vacía)")
            return
        newNode = Nodo(data)
        newNode.siguiente = self.inicio
        self.inicio.anterior = newNode
        self.inicio = newNode
        print("Nodo insertado al inicio")

    def insertEndList(self, data):
        if self.inicio is None:
            newNode = Nodo(data)
            self.inicio = newNode
            return
        n = self.inicio
        while n.siguiente is not None:
            n = n.siguiente
        newNode = Nodo(data)
        n.siguiente = newNode
        newNode.anterior = n
        print("Nodo insertado al final")

    def insertAfterNodo(self, X, data):
        if self.inicio is None:
            print("La lista está vacía")
            return
        n = self.inicio
        while n is not None:
            if n.dato == X:
                break
            n = n.siguiente
        if n is None:
            print("El elemento no existe en la lista")
        else:
            newNode = Nodo(data)
            newNode.anterior = n
            newNode.siguiente = n.siguiente
            if n.siguiente is not None:
                n.siguiente.anterior = newNode
            n.siguiente = newNode
            print(f"Nodo insertado después de {X}")

    def insertBeforeNodo(self, X, data):
        if self.inicio is None:
            print("La lista está vacía")
            return
        n = self.inicio
        while n is not None:
            if n.dato == X:
                break
            n = n.siguiente
        if n is None:
            print("El elemento no existe en la lista")
        else:
            newNode = Nodo(data)
            newNode.siguiente = n
            newNode.anterior = n.anterior
            if n.anterior is not None:
                n.anterior.siguiente = newNode
            else:
                # Si n es el primer nodo, actualizar inicio
                self.inicio = newNode
            n.anterior = newNode
            print(f"Nodo insertado antes de {X}")

    def imprimirLista(self):
        if self.inicio is None:
            print("La lista está vacía")
            return
        n = self.inicio
        print("Lista en orden: ", end="")
        while n is not None:
            print(n.dato, end=" <-> ")
            n = n.siguiente
        print("None")


# Ejemplo de uso
if __name__ == "__main__":
    lista = DoublyLinkedList()
    lista.insertEmptyList(10)
    lista.insertEndList(20)
    lista.insertBeginList(5)
    lista.insertAfterNodo(10, 15)
    lista.insertBeforeNodo(20, 18)
    lista.imprimirLista()
```

```python

# Desarrollar una función que permita buscar un elemento en una lista circular

class Nodo:
    def __init__(self, valor):
        self.valor = valor
        self.next = None


class ListaCircular:
    def __init__(self):
        self.inicio = None

    # Verifica si la lista está vacía
    def vacia(self):
        return self.inicio is None

    # Inserta un nodo al final de la lista
    def insertar(self, valor):
        nuevo = Nodo(valor)
        if self.vacia():
            self.inicio = nuevo
            nuevo.next = nuevo  # Se apunta a sí mismo porque es circular
        else:
            temp = self.inicio
            # Recorre hasta el último nodo
            while temp.next != self.inicio:
                temp = temp.next
            temp.next = nuevo
            nuevo.next = self.inicio
        print(f"Nodo {valor} insertado con éxito.")

    # Muestra los elementos de la lista circular
    def mostrar(self):
        if self.vacia():
            print(" La lista está vacía")
            return
        temp = self.inicio
        print("Lista circular:", end=" ")
        while True:
            print(temp.valor, end=" -> ")
            temp = temp.next
            if temp == self.inicio:
                break
        print("(vuelve al inicio)")

    # Busca un elemento en la lista circular
    def buscar(self, valor):
        if self.vacia():
            return False
        temp = self.inicio
        while True:
            if temp.valor == valor:
                return True   # encontrado
            temp = temp.next
            if temp == self.inicio:
                break
        return False  # no encontrado


# -------------------------------
# Programa principal interactivo
# -------------------------------
if __name__ == "__main__":
    lista = ListaCircular()

    while True:
        print("\n--- MENÚ LISTA CIRCULAR ---")
        print("1. Insertar un elemento")
        print("2. Mostrar lista")
        print("3. Buscar un elemento")
        print("4. Salir")

        opcion = input("Elige una opción: ")

        if opcion == "1":
            dato = input("Ingrese el valor a insertar: ")
            lista.insertar(dato)

        elif opcion == "2":
            lista.mostrar()

        elif opcion == "3":
            dato = input("Ingrese el valor a buscar: ")
            if lista.buscar(dato):
                print(f" El elemento {dato} SÍ está en la lista.")
            else:
                print(f" El elemento {dato} NO se encuentra en la lista.")

        elif opcion == "4":
            print(" Saliendo del programa...")
            break

        else:
            print(" Opción inválida, intenta nuevamente.")
```

```python
# Desarrollar una función que permita buscar un elemento en una lista Circular doblemente enlazada.

class NodoDoble:
    def __init__(self, valor):
        self.valor = valor
        self.next = None
        self.prev = None

class ListaCircularDoble:
    def __init__(self):
        self.head = None

    def esta_vacia(self):
        return self.head is None

    def insertar_final(self, valor):
        nuevo = NodoDoble(valor)
        if self.esta_vacia():
            self.head = nuevo
            nuevo.next = nuevo
            nuevo.prev = nuevo
        else:
            tail = self.head.prev
            tail.next = nuevo
            nuevo.prev = tail
            nuevo.next = self.head
            self.head.prev = nuevo

    def recorrer(self):
        if self.esta_vacia():
            print("Lista vacía")
            return
        actual = self.head
        while True:
            print(actual.valor, end=" <-> ")
            actual = actual.next
            if actual == self.head:
                break
        print("(circular)")

    #  Función de búsqueda
    def buscar(self, valor):
        if self.esta_vacia():
            return None

        actual = self.head
        posicion = 0
        while True:
            if actual.valor == valor:
                return posicion   # devuelve la posición en la lista
            actual = actual.next
            posicion += 1
            if actual == self.head:  # recorrido completo
                break
        return -1  # no encontrado


# -------------------------------
# Programa principal interactivo
# -------------------------------
if __name__ == "__main__":
    lista = ListaCircularDoble()

    while True:
        print("\n--- MENÚ LISTA CIRCULAR DOBLE ---")
        print("1. Insertar un elemento")
        print("2. Mostrar lista")
        print("3. Buscar un elemento")
        print("4. Salir")

        opcion = input("Elige una opción: ")

        if opcion == "1":
            dato = input("Ingrese el valor a insertar: ")
            lista.insertar_final(dato)

        elif opcion == "2":
            lista.recorrer()

        elif opcion == "3":
            dato = input("Ingrese el valor a buscar: ")
            posicion = lista.buscar(dato)
            if posicion != -1 and posicion is not None:
                print(f" El elemento {dato} está en la posición {posicion}.")
            else:
                print(f" El elemento {dato} NO se encuentra en la lista.")

        elif opcion == "4":
            print(" Saliendo del programa...")
            break

        else:
            print(" Opción inválida, intenta nuevamente.")
```

```python
# Desarrollar una función que permita eliminar un elemento por dato en una lista circular doblemente enlazada.

class NodoDoble:
    def __init__(self, valor):
        self.valor = valor
        self.next = None
        self.prev = None

class ListaCircularDoble:
    def __init__(self):
        self.head = None

    def vacia(self):
        return self.head is None

    def insertar(self, valor):
        nuevo = NodoDoble(valor)
        if self.vacia():
            self.head = nuevo
            nuevo.next = nuevo
            nuevo.prev = nuevo
        else:
            tail = self.head.prev
            tail.next = nuevo
            nuevo.prev = tail
            nuevo.next = self.head
            self.head.prev = nuevo

    def mostrar(self):
        if self.vacia():
            print("[]")
            return
        actual = self.head
        salida = []
        while True:
            salida.append(str(actual.valor))
            actual = actual.next
            if actual == self.head:
                break
        print(" <-> ".join(salida) + " (circular)")

    #  Eliminar por valor
    def eliminar(self, valor):
        if self.vacia():
            print(" No se puede eliminar, lista vacía")
            return False

        actual = self.head
        while True:
            if actual.valor == valor:
                # Caso: único nodo
                if actual.next == actual and actual.prev == actual:
                    self.head = None
                else:
                    # Reconectar enlaces
                    actual.prev.next = actual.next
                    actual.next.prev = actual.prev
                    # Caso: si eliminamos la cabeza, moverla
                    if actual == self.head:
                        self.head = actual.next
                print(f" Elemento {valor} eliminado")
                return True

            actual = actual.next
            if actual == self.head:  # recorrido completo
                break

        print(f"✘ Elemento {valor} no encontrado")
        return False


# -------------------------------
# Programa principal interactivo
# -------------------------------
if __name__ == "__main__":
    lista = ListaCircularDoble()

    while True:
        print("\n--- MENÚ LISTA CIRCULAR DOBLEMENTE ENLAZADA ---")
        print("1. Insertar un elemento")
        print("2. Mostrar lista")
        print("3. Eliminar un elemento")
        print("4. Salir")

        opcion = input("Elige una opción: ")

        if opcion == "1":
            dato = input("Ingrese el valor a insertar: ")
            lista.insertar(dato)

        elif opcion == "2":
            lista.mostrar()

        elif opcion == "3":
            dato = input("Ingrese el valor a eliminar: ")
            lista.eliminar(dato)

        elif opcion == "4":
            print(" Saliendo del programa...")
            break

        else:
            print(" Opción inválida, intenta nuevamente.")
```
